```python
import asyncio
from notte.env import NotteEnv
from examples.agent import SimpleNotteAgent
from examples.credentials import get_credentials

# For Jupyter notebooks, use await directly
username, password = get_credentials()

agent = SimpleNotteAgent(
    model="openai/gpt-4o-mini",
    max_steps=10,
    headless=False
)

# Run the agent directly with await
task = "Post 'Hello World' on Twitter"
result = await agent.run(
    task=task,
    url="https://twitter.com",
    credentials=username,
    password=password
)
print(result)
```
